```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df_patient = pd.read_csv("/content/patient_medication_data.csv")
df_patient.head()
```

```python
df_medication = pd.read_csv("/content/medication_data.csv")
df_medication.head()
```

```python
 #Rename the 'medicine_issued' column in df_patient to 'medicine'
df_patient = df_patient.rename(columns={'medicine_issued': 'medicine'})
```

```python
# Print the columns of both DataFrames to check their names
print("Columns in df_patient:", df_patient.columns)
print("Columns in df_medication:", df_medication.columns)
```

```python
# Rename the 'medicine' column in df_patient to 'medication' for consistency
df_patient = df_patient.rename(columns={'medicine': 'medication'})
```

```python
# Print the columns of both DataFrames to check their names
print("Columns in df_patient:", df_patient.columns)
print("Columns in df_medication:", df_medication.columns)
```

```python
# Perform an inner merge based on both 'medication' and 'diagnosis'
merged_df_both = pd.merge(df_patient, df_medication, on=['medication', 'diagnosis'], how='inner')
```

```python
# 1. Distribution of Diagnoses
plt.figure(figsize=(10, 6))
sns.countplot(y='diagnosis', data=merged_df_both, order=merged_df_both['diagnosis'].value_counts().index)
plt.title('Distribution of Diagnoses')
plt.xlabel('Number of Patients')
plt.ylabel('Diagnosis')
plt.show()
```

```python
# 2. Distribution of Medications
plt.figure(figsize=(12, 6))
sns.countplot(y='medication', data=merged_df_both, order=merged_df_both['medication'].value_counts().index)
plt.title('Distribution of Medications Prescribed')
plt.xlabel('Number of Prescriptions')
plt.ylabel('Medication')
plt.show()
```

```python
# 3. Medication Distribution by Gender
plt.figure(figsize=(12, 6))
sns.countplot(y='medication', hue='gender', data=merged_df_both, order=merged_df_both['medication'].value_counts().index)
plt.title('Medication Distribution by Gender')
plt.xlabel('Number of Prescriptions')
plt.ylabel('Medication')
plt.legend(title='Gender')
plt.show()
```

```python
# 4. Diagnosis Distribution by Gender
plt.figure(figsize=(10, 6))
sns.countplot(y='diagnosis', hue='gender', data=merged_df_both, order=merged_df_both['diagnosis'].value_counts().index)
plt.title('Diagnosis Distribution by Gender')
plt.xlabel('Number of Patients')
plt.ylabel('Diagnosis')
plt.legend(title='Gender')
plt.show()
```

```python
 #Visualization 1: Diagnoses distribution
plt.figure(figsize=(12, 6))
diagnosis_counts = merged_df_both['diagnosis'].value_counts()
sns.barplot(x=diagnosis_counts.index, y=diagnosis_counts.values)
plt.title('Distribution of Diagnoses')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('diagnosis_distribution.png')
plt.show()
```

```python
# Visualization 2: Gender distribution by diagnosis
plt.figure(figsize=(12, 6))
gender_diagnosis = pd.crosstab(merged_df_both['diagnosis'], merged_df_both['gender'])
gender_diagnosis.plot(kind='bar', stacked=True)
plt.title('Gender Distribution by Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Gender')
plt.tight_layout()
plt.savefig('gender_by_diagnosis.png')
plt.show()
```

```python
def recommend_substitute(medication, diagnosis, merged_df):
    """
    Recommends a potential drug substitute based on the provided medication and diagnosis.
    It looks up the 'drug_substitute' in the merged DataFrame where the
    'medication' and 'diagnosis' match.
    """
    recommendations = merged_df[(merged_df['medication'] == medication) & (merged_df['diagnosis'] == diagnosis)]['drug_substitute'].unique()
    if len(recommendations) > 0:
        return recommendations
    else:
        return f"No substitute found in the data for {medication} with diagnosis {diagnosis}."
```

```python
medication_input = 'Lisinopril'
diagnosis_input = 'Hypertension'
recommendations = recommend_substitute(medication_input, diagnosis_input, merged_df_both)
print(f"\nFor medication '{medication_input}' and diagnosis '{diagnosis_input}', potential substitutes are: {recommendations}")
```
